```python
df['month']=df['Order Date'].dt.month
df['year']=df['Order Date'].dt.year
df.head()
```

```python
#profit margin
df['profit_margin']=round(df.Profit/df.Sales,ndigits=2)
df.head()
```

```python
#total profit and sales by sub category
df.groupby('Sub-Category').sum()[['Sales','Profit']].plot(kind='bar')
print(df.groupby('Sub-Category').sum()[['Sales','Profit']])
```

```python
# define a function to return a column name aggregated by sum and the index reset
def group(col):
        
        a=df.groupby(col).sum().reset_index()
        return a
```

```python
group('Segment')
```

```python
px.histogram(df.Sales)
```

```python
#sales by sub category
sub=group('Sub-Category')
px.bar(sub,x='Sub-Category',y='Sales')
```

```python
px.bar(sub,x='Sub-Category',y='Profit')
```

```python
'plot sales and profit by sub category'
px.bar(sub,x='Sub-Category',y='Sales',color='Profit',color_continuous_scale=['red','yellow','green'],text_auto='.2s')
```

```python
#inspect negative profit of tables
#is there a correlation between sales/profit and discount?
px.scatter(df,x='Sales',y='Profit',color='Discount',hover_name='Sub-Category')
#higher discount leads to less profit
```

```python
discount=df.groupby('Sub-Category').agg({'Discount':'mean','Profit':'sum'})
discount.head()
```

```python
px.bar(discount,x=discount.index,y='Discount',color='Profit',color_continuous_scale=['red','yellow','green'])
#in most casesa lower discount yields a higher profit. an exception can be seen with binders
```
