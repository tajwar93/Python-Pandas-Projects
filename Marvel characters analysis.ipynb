```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px

sns.set_style('whitegrid')
plt.rcParams['font.size']=12
plt.rcParams['figure.figsize']=(15,5)
plt.rcParams['figure.constrained_layout.use'] = True

df1=pd.read_csv(r'C:\Users\TajwarAbtahee\OneDrive - JCW Resourcing\Desktop\Python\practice\charcters_stats.csv')
df2=pd.read_csv(r'C:\Users\TajwarAbtahee\OneDrive - JCW Resourcing\Desktop\Python\practice\marvel_characters_info.csv')

df1
```

```python
df2
```

```python
df=df1.merge(df2,how='inner',on='Name')
df
```

```python
df=df[df.Publisher=='Marvel Comics']
df.drop(columns=['SkinColor','ID','Alignment_y'],inplace=True)
```

```python
df.corr().style.background_gradient(cmap='RdYlGn')
```

```python
fig,axes=plt.subplots(nrows=5,figsize=(15,20))

sns.countplot(df.Alignment_x,ax=axes[0])
sns.countplot(df.Gender,ax=axes[1])
df.EyeColor.value_counts().plot(kind='bar',ax=axes[2],xlabel='Eye Color')
df.Race.value_counts().plot(kind='bar',ax=axes[3],xlabel='Race')
df.HairColor.value_counts().plot(kind='bar',ax=axes[4],xlabel='Hair Color')

plt.show()

#most popular counts in each category
```

```python
sns.distplot(df.Intelligence,label='Intelligence')
sns.distplot(df.Strength,label='Strength')
sns.distplot(df.Speed,label='Speed')
sns.distplot(df.Durability,label='Durability')
sns.distplot(df.Power,label='Power')
sns.distplot(df.Combat,label='Combat')
plt.legend()

plt.show()

#strength and speed are closer together and also higher in each group
# intelligence and power are also close together and seenm to go hand in hand
```

```python
fig,axes=plt.subplots(nrows=5)

sns.boxplot(df.Intelligence,ax=axes[0])
sns.boxplot(df.Strength,ax=axes[1])
sns.boxplot(df.Speed,ax=axes[2])
sns.boxplot(df.Durability,ax=axes[3])
sns.boxplot(df.Power,ax=axes[4])
sns.boxplot(df.Combat,ax=axes[5])

plt.show()
```

```python
good=df[df.Alignment_x=='good']
bad=df[df.Alignment_x=='bad']

fig,axes=plt.subplots(ncols=2,nrows=7,figsize=(15,20))

good.sort_values('Intelligence',ascending=False).head(20).plot(kind='bar',x='Name',y='Intelligence',ax=axes[0,0])
bad.sort_values('Intelligence',ascending=False).head(20).plot(kind='bar',x='Name',y='Intelligence',color='r',ax=axes[0,1])

good.sort_values('Strength',ascending=False).head(20).plot(kind='bar',x='Name',y='Strength',ax=axes[1,0])
bad.sort_values('Strength',ascending=False).head(20).plot(kind='bar',x='Name',y='Strength',color='r',ax=axes[1,1])

good.sort_values('Speed',ascending=False).head(20).plot(kind='bar',x='Name',y='Speed',ax=axes[2,0])
bad.sort_values('Speed',ascending=False).head(20).plot(kind='bar',x='Name',y='Speed',color='r',ax=axes[2,1])

good.sort_values('Power',ascending=False).head(20).plot(kind='bar',x='Name',y='Power',ax=axes[3,0])
bad.sort_values('Power',ascending=False).head(20).plot(kind='bar',x='Name',y='Power',color='r',ax=axes[3,1])

good.sort_values('Combat',ascending=False).head(20).plot(kind='bar',x='Name',y='Combat',ax=axes[4,0])
bad.sort_values('Combat',ascending=False).head(20).plot(kind='bar',x='Name',y='Combat',color='r',ax=axes[4,1])

good.sort_values('Durability',ascending=False).head(20).plot(kind='bar',x='Name',y='Durability',ax=axes[5,0])
bad.sort_values('Durability',ascending=False).head(20).plot(kind='bar',x='Name',y='Durability',color='r',ax=axes[5,1])

good.sort_values('Total',ascending=False).head(20).plot(kind='bar',x='Name',y='Total',ax=axes[6,0])
bad.sort_values('Total',ascending=False).head(20).plot(kind='bar',x='Name',y='Total',color='r',ax=axes[6,1])

plt.show()

#most powerful heroes and villains in each stat category
```
