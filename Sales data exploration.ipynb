```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df=pd.read_excel(r'C:\Users\TajwarAbtahee\OneDrive - JCW Resourcing\Desktop\Python\superstore_sales.xlsx')
df.head() #first 5 rowa
```

```python
df.shape #total rows and columns
```

```python
df.info() #summarized column info. there are no null values
```

```python
#show the sales trend on a month/year basis

df['Month_year']=df.order_date.dt.strftime('%Y-%m')
df
```

```python
df.groupby('Month_year').sum()
```

```python
trend=df.groupby('Month_year').sum().sales.reset_index()
trend
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=trend,x=trend.Month_year,y=trend.sales)
plt.xticks(rotation=90, size=16)
plt.yticks(size=16)
plt.show()
```

```python
#what are the top 10  products by sales?
df.groupby(['product_name']).sum()['sales']
```

```python
pd.DataFrame(df.groupby(['product_name']).sum()['sales']).sort_values('sales',ascending=False).head(10)
```

```python
#what are the top 5 most selling products?
pd.DataFrame(df.groupby('product_name').sum().quantity).sort_values('quantity', ascending=False).head(5)
```

```python
b=df.ship_mode.value_counts().reset_index()
b
```

```python
#what is the most preferred ship mode?
plt.figure(figsize=(20,10))
a=sns.countplot(df.ship_mode)
a.bar_label(container=a.containers[0],labels=b.ship_mode, size=19)
plt.xticks(size=16)
plt.yticks(size=16)
plt.xlabel('ship mode',size=16)
plt.ylabel('count',size=16)
plt.show()
```

```python
#what is the most profitable category and sub-category?
df.head()
profit=pd.DataFrame(df.groupby('category').sum().profit.round(decimals=2)).sort_values('profit',ascending=False)
profit
```

```python
plt.figure(figsize=(20,10))
c=sns.barplot(x=profit.index,y=profit.profit,data=profit)
c.bar_label(container=c.containers[0],labels=profit.profit,size=16)
plt.xticks(size=16)
plt.yticks(size=16)
plt.xlabel('Category',size=16)
plt.ylabel('Profit in Â£',size=16)
plt.show()
```
