```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px

sns.set_style('whitegrid')
plt.rcParams['font.size']=12
plt.rcParams['figure.figsize']=(15,5)

df=pd.read_csv(r'C:\Users\TajwarAbtahee\OneDrive - JCW Resourcing\Desktop\Python\practice\StudentsPerformance.csv')
df.head()
```

```python
df.info()
```

```python
a=sns.countplot(df.gender)
a.bar_label(container=a.containers[0],labels=df.gender.value_counts())
plt.show()

#a close number of males and females
```

```python
race=df['race/ethnicity'].value_counts().to_frame()
a=sns.barplot(data=race,x=race.index,y='race/ethnicity')
a.bar_label(container=a.containers[0],labels=race['race/ethnicity'])
plt.show()
```

```python
edu=df['parental level of education'].value_counts().to_frame()
b=sns.barplot(data=edu,x=edu.index,y='parental level of education')
b.bar_label(container=b.containers[0],labels=edu['parental level of education'])
plt.show()
```

```python
lunch=df['lunch'].value_counts().to_frame()
c=sns.barplot(data=lunch,x=lunch.index,y='lunch')
c.bar_label(container=c.containers[0],labels=lunch['lunch'])
plt.show()
```

```python
prep=df['test preparation course'].value_counts().to_frame()
d=sns.barplot(data=prep,x=prep.index,y='test preparation course')
d.bar_label(container=d.containers[0],labels=prep['test preparation course'])
plt.show()
```

```python
fig,axes=plt.subplots()
sns.kdeplot(df['math score'],shade=True,axes=axes)
sns.kdeplot(df['reading score'],shade=True,axes=axes)
sns.kdeplot(df['writing score'],shade=True,axes=axes)
plt.legend(labels=["math score","reading score","writing score"])
plt.xlabel('subject scores')
plt.show()
#most students have a score between 60-80 for the subjects
#lower scores in maths compared to other
#reading and writing are very close and this goes to show that a person reading level is in proportion to their writing level and vice versa
```

```python
sns.pairplot(df,vars=['math score','reading score','writing score'],hue='gender')
#strong positive correllations across the board
#females tend to do better than males in all comparisons
```

```python
sns.pairplot(df,vars=['math score','reading score','writing score'],hue='race/ethnicity',size=5)
# alot of group c are clustered in the middle
# group E and D have results near the top across the board
# group B and C have a lower score for a few of their entiries across the board
```

```python
sns.pairplot(df,vars=['math score','reading score','writing score'],hue='parental level of education', size=5)
#those with higher levels of education have performed better
```

```python
sns.pairplot(df,vars=['math score','reading score','writing score'],hue='lunch',size=5)
# those with standard lunches performed far better than their free lunch counterparts
```

```python
sns.pairplot(df,vars=['math score','reading score','writing score'],hue='test preparation course',size=5)
#those who completed the test prep course performed better
```

```python
fig, axes=plt.subplots(nrows=3,figsize=(10,10))
sns.boxplot(df['math score'],ax=axes[0])
sns.boxplot(df['reading score'],ax=axes[1])
sns.boxplot(df['writing score'],ax=axes[2])
plt.show()

#the data across the board seems similar
# math scores are more consistent but have the most outliers are smallest IQR
# reading and writing are very similar, although writing is slightly more consistent
```
