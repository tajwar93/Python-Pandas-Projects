```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px

sns.set_style('whitegrid')
plt.rcParams['font.size']=12
plt.rcParams['figure.figsize']=(15,5)
plt.rcParams['figure.constrained_layout.use'] = True

df=pd.read_csv(r'C:\Users\TajwarAbtahee\OneDrive - JCW Resourcing\Desktop\Python\practice\superheroes_power_matrix.csv')
df
```

```python
df.replace({True: 1, False: 0},inplace=True)
```

```python
df['total']=df.iloc[:,1:].sum(axis=1)
df
```

```python
tots=df.sort_values('total',ascending=False).head(20)
```

```python
df.head()
```

```python
def name(x):
    a=df[df.Name.isin([x])]
    b=a[a==1].dropna(axis=1).columns
    return b

#this function returns a character with all their abilities
```

```python
name('Yoda')
```

```python
name('Flash')
```

```python
name('Apocalypse')
```

```python
sns.barplot(data=tots,x='Name',y='total')
plt.xticks(rotation=45)
plt.show()
print(tots[['Name','total']])
#characters with most abilites top 20
```

```python
sns.barplot(data=df.sort_values('total',ascending=False).tail(20),x='Name',y='total')
plt.xticks(rotation=45)
plt.show()

#bottom 20 characets only have 1 ability
```
