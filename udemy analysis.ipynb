```python
rel=df[['price', 'num_subscribers',
       'num_reviews', 'num_lectures']]
rel.corr().style.background_gradient(cmap='RdYlGn')
```

```python
#top selling coures
sell=df.sort_values('num_subscribers',ascending=False).head()
plt.figure(figsize=(14,4))
ax=sns.barplot(data=sell,x='course_title',y='num_subscribers')
ax.bar_label(container=ax.containers[0],labels=sell.num_subscribers)
plt.title('top selling courses')
plt.xticks(rotation='vertical')
plt.show()
```

```python
sub=df.groupby('subject').price.mean().reset_index()
sub=sub.sort_values('price',ascending=False)
px.bar(sub,x='subject',y='price',text_auto='.2s',title='Average price per subject')
```

```python
px.histogram(df, x="subject", color="is_paid")
#web dev has one of the most paid courses aswel as having the most free courses
```

```python
sub2=df.groupby('subject').num_subscribers.mean().reset_index()
sub2=sub2.sort_values('num_subscribers',ascending=False)
px.bar(sub2,x='subject',y='num_subscribers',text_auto='.2s',title='Average subscibers per subject')

#web development is clearly the most popular course
```
