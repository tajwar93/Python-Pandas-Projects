```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px

df=pd.read_csv(r'C:\Users\TajwarAbtahee\OneDrive - JCW Resourcing\Desktop\Python\practice\Marketing_Analysis.csv')
df.head()
```

```python
df=pd.read_csv(r'C:\Users\TajwarAbtahee\OneDrive - JCW Resourcing\Desktop\Python\practice\Marketing_Analysis.csv',skiprows=2)
df.head()
```

```python
df.drop(columns='customerid',inplace=True)
df.head()
#drop customerid as it is unecessary
```

```python
#from jobedu extract the job and level of educaition
df[['job','edu']]=df.jobedu.str.split(',',expand=True)
```

```python
df.head()
```

```python
#now lets drop jobedu
df.drop(columns='jobedu',inplace=True)
df.head()
```

```python
df
```

```python
df.isnull().sum()
```

```python
#drop columns that have no age
df=df[~df.age.isnull()]
```

```python
df.isnull().sum()
```

```python
#find the mode of the month
month_mode=df.month.mode()[0]
```

```python
#fill missing values in the moth column with the modal month
df.month.fillna(month_mode,inplace=True)
df.isnull().sum()
```

```python
#drop rows with no response
df=df[~df.response.isnull()]
```

```python
#what is the percentage of each job type?
df.job.value_counts(normalize=True)
```

```python
df.job.value_counts(normalize=True).plot(kind='barh')
plt.show()
#most people are either in blue collar jobs or management
```

```python
#percentage of each level of education
df.edu.value_counts(normalize=True)
```

```python
df.edu.value_counts(normalize=True).plot(kind='barh')
plt.show()
```

```python
#above data on a pie chart
df.edu.value_counts(normalize=True).plot(kind='pie')
#more than half of the people have compelted secondary eduation, with more than a quarter in tertiary
```

```python
#lets look at the salary stats
df.salary.describe()
#we have an average salary of around 57,000
#there seems to be data that tells us people are working for free
#the max salary looks like an outlier
```

```python
#lets compare relationships between age, salary and balance(peoples savings)
px.scatter(df,x='salary',y='balance')
#with a salary between 0-20k people have roughly the same as their savings with outliers increasing as salary increases
#most people have a salary of 100k but savings seem to be less than their salary
```

```python
px.scatter(df,x='age',y='balance')
#outliers start to diverge after 40k+
#people have the most money saved between 30-60 years old
```

```python
#lets check the corellation between the stated variables
df[['age','salary','balance']].corr().style.background_gradient(cmap='RdYlGn')
#although correlation is weak, age vs balance is the highest and age vs salary is the lowest
```

```python
#compare mean salaries of peoples responses of yes and no
df.groupby('response').mean()['salary']
#peope that answered yes had a salary £2000 higher than those that said no
```

```python
#lets look at the above data in a boxplot
sns.boxplot(data=df,x='response',y='salary')
plt.show()
# for most people that said no, have a salary between 20-70k, but 50-100k for people that said yes
#the median value is equal for both as well as the range
```

```python
df.response.value_counts()
```

```python
#lets convert the yes and no into 1 and 0
import numpy as np
df['response rate']=np.where(df.response=='yes',1,0)
df['response rate'].value_counts()
```

```python
#lets check the marital status of the people that responded
df.groupby('marital').mean()['response rate'].plot(kind='bar')
#most people that took part were singles
```

```python
df.groupby('loan').mean()['response rate'].plot(kind='bar')
#people that responded as no have more than double the loans compared to their yes counterparts
```

```python
result=pd.pivot_table(data=df,index='edu',columns='marital',values='response rate')
print(result)
sns.heatmap(result,annot=True,cmap='RdYlGn')
plt.show()
#single people with tertiary education responded best, and married people with primary education responded worst
```
