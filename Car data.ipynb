```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px

sns.set_style('whitegrid')
plt.rcParams['font.size']=12
plt.rcParams['figure.figsize']=(15,5)

df=pd.read_csv(r'C:\Users\TajwarAbtahee\OneDrive - JCW Resourcing\Desktop\Python\practice\Car_sales.csv')
df.head()
```

```python
df.info()
```

```python
sns.heatmap(df.isnull())
```

```python
df['Price_in_thousands'] = df['Price_in_thousands'].fillna(df.groupby('Manufacturer')['Price_in_thousands'].transform('mean'))
df['Price_in_thousands']=round(df['Price_in_thousands'],ndigits=2)

df['__year_resale_value'] = df['__year_resale_value'].fillna(df.groupby('Manufacturer')['__year_resale_value'].transform('mean'))
df['__year_resale_value']=round(df['__year_resale_value'],ndigits=2)

df['Engine_size'] = df['Engine_size'].fillna(df.groupby('Manufacturer')['Engine_size'].transform('mean'))
df['Engine_size']=round(df['Engine_size'],ndigits=2)

df['Horsepower'] = df['Horsepower'].fillna(df.groupby('Manufacturer')['Horsepower'].transform('mean'))
df['Horsepower']=round(df['Horsepower'],ndigits=2)

df['Wheelbase'] = df['Wheelbase'].fillna(df.groupby('Manufacturer')['Wheelbase'].transform('mean'))
df['Wheelbase']=round(df['Wheelbase'],ndigits=2)

df['Width'] = df['Width'].fillna(df.groupby('Manufacturer')['Width'].transform('mean'))
df['Width']=round(df['Width'],ndigits=2)

df['Length'] = df['Length'].fillna(df.groupby('Manufacturer')['Length'].transform('mean'))
df['Length']=round(df['Length'],ndigits=2)

df['Curb_weight'] = df['Curb_weight'].fillna(df.groupby('Manufacturer')['Curb_weight'].transform('mean'))
df['Curb_weight']=round(df['Curb_weight'],ndigits=2)

df['Fuel_capacity'] = df['Fuel_capacity'].fillna(df.groupby('Manufacturer')['Fuel_capacity'].transform('mean'))
df['Fuel_capacity']=round(df['Fuel_capacity'],ndigits=2)

df['Fuel_efficiency'] = df['Fuel_efficiency'].fillna(df.groupby('Manufacturer')['Fuel_efficiency'].transform('mean'))
df['Fuel_efficiency']=round(df['Fuel_efficiency'],ndigits=2)

df['Power_perf_factor'] = df['Power_perf_factor'].fillna(df.groupby('Manufacturer')['Power_perf_factor'].transform('mean'))
df['Power_perf_factor']=round(df['Power_perf_factor'],ndigits=2)

df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df.corr().style.background_gradient(cmap='RdYlGn')
```

```python
df.rename(columns={'__year_resale_value':'year_r_val'},inplace=True)
```

```python
sns.distplot(df.Sales_in_thousands,label='Sales in 1000s',bins=40)
sns.distplot(df.year_r_val,label='resale value in 1000s',bins=40)
plt.legend()
plt.show()
#data is right skewed so lets apply a log transformations to eliminate outliers from visualisaiton
```

```python
import numpy as np
sns.distplot(np.log1p(df.Price_in_thousands),label='Price in 1000s',bins=40)
sns.distplot(np.log1p(df.year_r_val),label='resale value in 1000s',bins=40)
plt.legend()
plt.show()
#most cars sell for around £3000
#whereas most cars resell for around £2700
```
